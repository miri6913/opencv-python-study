# OpenCV Mini-Projects (imgproc Practice)

A list of small, self-contained projects to practice the topics from the [OpenCV imgproc tutorials](https://docs.opencv.org/4.x/d7/da8/tutorial_table_of_content_imgproc.html). Each project lists the tutorial sections it reinforces and a short brief on what to build. Try them in order — they get progressively harder.

---

## 1. Photo Booth Filters
**Practices:** Smoothing Images, Adding borders, Basic Drawing, Random generator and text

Build a script that loads any photo and outputs four versions side-by-side: original, blurred (Gaussian), median-filtered (good for noise), and bilateral-filtered (preserves edges). Add a colored border around each, label each one with `cv.putText`, and save the grid as a single image.

*Stretch:* Add a slider (`cv.createTrackbar`) to control the blur kernel size in real time.

---

## 2. Pencil Sketch Generator
**Practices:** Smoothing, Thresholding, Canny Edge Detector

Convert a photo to a pencil-sketch effect. Pipeline: grayscale → invert → Gaussian blur → blend back with the grayscale via dodge (`cv.divide`). Then make a second version that uses Canny edges on a smoothed image and inverts it. Compare which looks better.

*Stretch:* Try the same with adaptive thresholding instead of Canny — see which gives nicer line art.

---

## 3. Document Scanner
**Practices:** Canny, Finding contours, Convex Hull, Bounding boxes, Thresholding

Take a photo of a piece of paper on a desk. Detect its four corners, warp it to a flat top-down view, and apply adaptive thresholding so it looks like a black-and-white scan.

Pipeline: blur → Canny → find contours → keep the largest 4-corner contour → `cv.getPerspectiveTransform` → `cv.warpPerspective` → `cv.adaptiveThreshold`.

*Stretch:* Run it on your webcam feed and overlay the detected corners live.

---

## 4. Coin Counter
**Practices:** Hough Circle Transform, Smoothing, Basic Drawing

Take a picture of coins on a flat surface. Use `cv.HoughCircles` to detect them, draw a circle and a number on each, and print the total count in the corner.

*Stretch:* Estimate denomination by clustering circle radii (e.g., the largest 25% are quarters).

---

## 5. Color-Based Object Tracker
**Practices:** Thresholding with inRange, Image Moments, Bounding boxes, Finding contours

Pick a brightly colored object (a sticky note, a ball). Open the webcam, convert each frame to HSV, use `cv.inRange` to mask the object, find the largest contour, compute its centroid via image moments, and draw a tracking dot that follows it. Keep the last 30 centroids and draw a fading trail.

*Stretch:* Add HSV trackbars so you can tune the color range without editing code.

---

## 6. Lane Line Detector
**Practices:** Canny, Hough Line Transform, Remapping (ROI mask)

Use a still image of a road (Google any "highway dashcam" image). Convert to grayscale, run Canny, mask everything outside a triangular ROI in the lower half of the frame, then run `cv.HoughLinesP`. Draw the detected lane lines in red on top of the original.

*Stretch:* Average the left-side and right-side lines into a single line per side and extrapolate them to the bottom of the frame.

---

## 7. Histogram Equalization Comparison
**Practices:** Histogram Calculation, Histogram Equalization, Histogram Comparison

Load a low-contrast photo (something taken indoors with bad lighting works well). Show three versions: original, after global `cv.equalizeHist`, after CLAHE (`cv.createCLAHE`). Plot the histogram of each below the image using `cv.calcHist` and either matplotlib or hand-drawn bars.

*Stretch:* Quantify how much each version differs from the original using `cv.compareHist` with several methods (correlation, chi-square, Bhattacharyya).

---

## 8. Template Matching Mini-Game
**Practices:** Template Matching, Basic Drawing

Take a screenshot of a "Where's Waldo"-style busy image and crop a small template out of it. Use `cv.matchTemplate` with several methods (`TM_CCOEFF_NORMED` etc.) to find it back in the original. Draw a rectangle on the match and display the confidence score.

*Stretch:* Find multiple matches by thresholding the result map and using non-max suppression.

---

## 9. Touching Object Counter (Watershed)
**Practices:** Distance Transform, Watershed, Morphology Operations, Thresholding

Take a photo of rice grains, beans, or coins where some are touching each other. A naive contour count will merge them. Use the watershed pipeline: Otsu threshold → `cv.morphologyEx` opening → distance transform → threshold for sure-foreground → subtract to get unknown region → `cv.watershed` → count distinct labels.

*Stretch:* Color each segmented object differently and overlay on the original.

---

## 10. Shape Classifier
**Practices:** Finding contours, Convex Hull, Bounding boxes/circles, Image Moments, Point Polygon Test

Draw triangles, rectangles, and circles in any paint app and save the image. Threshold it, find each contour, and classify the shape based on the number of vertices in its `cv.approxPolyDP` result. Label each shape on the image and print a tally.

*Stretch:* Add a "is this point inside?" tester — click on the image and use `cv.pointPolygonTest` to report which shape (if any) you clicked.

---

## Suggested order to attempt
Start with **1 → 2 → 4** for warm-up (filters and Hough). Then do **5 → 7** for color and histograms. Then tackle **3 → 6 → 8** which combine multiple operations. Save **9 → 10** for last; they put everything together.

## Tips while practicing
- Always look at intermediate steps. `cv.imshow` your mask, your Canny output, your distance transform — debugging blindly is painful.
- Press `q` to quit webcam loops, and remember `cv.waitKey(1) & 0xFF`.
- When a result looks wrong, 90% of the time it's the threshold or kernel size. Wire up trackbars early.
- Keep a `utils.py` with helpers like `show(img, title)` and `stack(imgs)` so you don't rewrite boilerplate every project.


# Other Project ideas

**Project 1: Color-Based Object Tracker (webcam or video)**
Point a camera at a colored object (a ball, a sticky note) and track it in real-time.

Covers: colorspace conversion (BGR→HSV), `inRange` thresholding, morphological operations (erode/dilate to clean up the mask), contour detection, bounding boxes. This is a great starter because you see immediate visual feedback.

---

**Project 2: Document Scanner**
Take a phone photo of a document on a desk → output a clean, flat, black-and-white scan.

Covers: Gaussian blur, Canny edge detection, contour finding + approximation (`approxPolyDP`), perspective warp (`getPerspectiveTransform` + `warpPerspective`), adaptive thresholding. This is the classic "chain everything together" project.

---

**Project 3: License Plate / Sign Region Detector**
Find and extract rectangular regions of interest from street-level photos (doesn't need OCR — just locating and cropping).

Covers: Sobel/Laplacian gradients, morphological closing (to merge nearby edges into blobs), contour filtering by area/aspect ratio, bounding rotated rectangles (`minAreaRect`), image pyramids for multi-scale detection. Forces you to tune parameters on messy real-world images, which is where the real learning happens.

---

**Project 4: Image Histogram Analyzer & Auto-Enhancer**
Load any photo → show its histogram → auto-correct exposure/contrast → show the before/after with the updated histogram.

Covers: `calcHist`, histogram equalization (`equalizeHist`), CLAHE (adaptive equalization), histogram comparison, back-projection. You'll also naturally use colorspace conversion again (work in LAB or YCrCb for better results than equalizing BGR channels directly).

---

**Project 5: Panorama Stitcher (simplified)**
Take 2–3 overlapping photos → stitch them into a panorama.

Covers: template matching, image transforms (affine/perspective), image pyramids for blending, border handling (`copyMakeBorder`), and combining masks. This is the capstone because it pulls together geometric transforms, filtering, and blending in a single pipeline.
