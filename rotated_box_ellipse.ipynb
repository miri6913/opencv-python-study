### Rotated box tasks
Task 1 — Draw rotated box around one object

Task 2 — Print object angle

Task 3 — Compare with boundingRect

Task 4 — Measure width and height of tilted object

```python
import cv2
import numpy as np

img = cv2.imread('Data/tilted_rectangle.png', cv2.COLOR_BGR2GRAY)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
cv2.imshow("thresh", thresh)


contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
rect = cv2.minAreaRect(contours[0])
print(rect)

box = cv2.boxPoints(rect)
print(box)

box = np.int0(box)
print(box)

cv2.drawContours(img, [box], 0, (0, 0, 255), 6)
cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(f"Angle: {rect[2]:.4f}")
```

```python
print(f"Width and height: {rect[1]}")
```

```python
x, y, w, h = cv2.boundingRect(contours[0])

cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 5)
cv2.imshow("brect", img)
cv2.waitKey()
cv2.destroyAllWindows()
```
